# Mini Project 1: Exploratory Data Analysis

## Introduction

This dataset represents customer shopping behavior and engagement patterns.
Each row corresponds to a unique customer identified by Customer_ID.

The dataset includes demographic information (Age, Gender), financial data (Annual Income), 
and behavioral indicators such as Spending Score, Online Purchases, Store Visits, 
and Membership Years.

The objective of this analysis is to explore customer behavior patterns, 
identify relationships between variables, and extract meaningful business insights.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
data = {
    "Customer_ID": [101,102,103,104,105,106,107,108,109,110,111,112,113,114,115],
    "Gender": ["Female","Male","Female","Male","Female","Male","Female","Male","Female","Male","Female","Male","Female","Male","Female"],
    "Age": [23,35,45,31,22,40,29,50,27,33,38,41,26,48,np.nan],
    "Annual_Income (k$)": [25,60,80,45,30,90,50,100,35,70,85,95,40,110,55],
    "Spending_Score (1-100)": [39,81,6,77,40,76,45,3,50,88,10,95,55,2,60],
    "Online_Purchases": [5,20,2,18,6,22,10,1,8,25,3,27,9,0,15],
    "Store_Visits_Per_Month": [4,2,1,3,5,2,4,1,6,3,2,2,5,1,4],
    "Membership_Years": [1,5,8,3,1,10,4,12,2,6,7,9,3,15,5]
}

df = pd.DataFrame(data)
df
```

```python
## Data Cleaning Decisions

1. One missing value was found in the Age column. Since only a single value was missing,
   it was replaced using median imputation to maintain dataset balance without
   significantly affecting the distribution.

2. No duplicate records were identified, ensuring data consistency.

3. No columns were dropped, as all variables provide meaningful behavioral
   or demographic information for analysis.

The dataset is now clean and reliable for exploratory analysis.
```

```python
### Insight: Age Distribution

Most customers fall between 25 and 45 years of age, indicating the business primarily
serves young to middle-aged individuals. This suggests marketing strategies may
be optimized for this age demographic.
```

```python
plt.figure()
plt.hist(df["Age"], bins=5)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.show()
```

```python
### Insight: Gender Distribution

The dataset shows a nearly balanced distribution of male and female customers.
This indicates the business appeals equally to both genders, allowing for
gender-neutral marketing strategies.
```

```python
plt.figure()
sns.countplot(x="Gender", data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
### Insight: Spending Behavior

Spending scores vary significantly across customers, showing diverse purchasing
behavior. This variation suggests potential customer segmentation opportunities
based on spending intensity.
```

```python
plt.figure()
sns.histplot(df["Spending_Score (1-100)"], kde=True)
plt.title("Spending Score Distribution")
plt.show()
```

```python
### Insight: Income vs Spending Score

There is no strong linear relationship between annual income and spending score.
This suggests that higher income does not necessarily translate into higher spending,
indicating that behavioral factors influence purchasing decisions.
```

```python
plt.figure()
sns.scatterplot(x="Annual_Income (k$)", y="Spending_Score (1-100)", data=df)
plt.title("Income vs Spending Score")
plt.show()
```

```python
## Multivariate Analysis Insight

The correlation heatmap reveals that Membership Years and Online Purchases
have a positive relationship, highlighting the importance of customer retention.

Age shows moderate correlation with Membership Years, suggesting older
customers may demonstrate greater loyalty.

Annual Income does not strongly correlate with Spending Score, reinforcing
that spending behavior depends more on engagement than income level.
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Correlation Matrix
correlation_matrix = df.corr(numeric_only=True)

print(correlation_matrix)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

# Executive Summary

1. Customer loyalty significantly impacts online purchasing behavior.
   Long-term members are more active in digital shopping, emphasizing
   the value of retention strategies.

2. Income does not directly determine spending patterns.
   This suggests psychological and behavioral factors influence purchases
   more than financial capacity alone.

3. Age appears moderately related to membership duration,
   indicating older customers may remain engaged with the business longer.

Overall, the dataset highlights clear opportunities for customer segmentation,
loyalty-based marketing, and personalized engagement strategies.
# Overall Project Summary

This analysis explored customer demographics, income levels, and purchasing behavior to understand key engagement patterns. After cleaning the dataset by handling a minor missing value and confirming no duplicates, exploratory analysis revealed that the business primarily serves young to middle-aged customers with balanced gender representation. 

Spending behavior varies significantly across customers and is not strongly influenced by income level. Instead, customer loyalty plays a more important role—members with longer membership durations tend to make more online purchases. 

Overall, the findings suggest that retention strategies and behavioral segmentation are more effective for business growth than income-based targeting.
